## Statistical Test
## 1. shapiro()

### Task
1. shapiro() them both and compare.
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html

##### Expected outcome (actual values may be different.)

stat_1: 0.9409, p_1: 0.0002

stat_2: 0.9887, p_2: 0.5616

For data_1 : 

Sample does not look Gaussian.

For data_2 : 

It can be said that it is normally distributed.

```python
stat_1, p_1=stat.shapiro(data_1)
print("stat_1: %.4f, p_1: %.4f"%(stat_1,p_1))
stat_2, p_2=stat.shapiro(data_2)
print("stat_2: %.4f, p_2: %.4f"%(stat_2,p_2))

alpha=0.05
print("\nFor data_1 : ")
if p_1 > alpha:
    print("It can be said that it is normally distributed.")
else:
    print("Sample does not look Gaussian.")

print("\nFor data_2 : ")
if p_2 > alpha:
    print("It can be said that it is normally distributed.")
else:
    print("Sample does not look Gaussian.")
```

## 2. normaltest()
### Task
1. normaltest() them both and compare. https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.normaltest.html

##### Expected outcome

stat_1: 49.4264, p_1: 0.0000

stat_2: 0.9779, p_2: 0.6133

For data_1:

Sample does not look Gaussian.

For data_2 : 

It can be said that it is normally distributed.

```python
stat_1, p_1 = stat.normaltest(data_1)
print("stat_1: %.4f, p_1: %.4f"%(stat_1,p_1))
stat_2, p_2 = stat.normaltest(data_2)
print("stat_2: %.4f, p_2: %.4f"%(stat_2,p_2))

alpha=0.05
print("\nFor data_1:")
if p_1 > alpha:
    print("It can be said that it is normally distributed.")
else:
    print("Sample does not look Gaussian.")

print("\nFor data_2 : ")
if p_2 > alpha:
    print("It can be said that it is normally distributed.")
else:
    print("Sample does not look Gaussian.")
```
