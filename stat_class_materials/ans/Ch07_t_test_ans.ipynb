## one-sample t-test
### Task
1. Given the following sample, test if its true mean is 1.75.
2. Remember, for sample stdev or variance, set ddof = 1.
3. Compute the t_value and p_value.
4. Plot it.
5. Make it two-sided.

```python
# Given
np.random.seed(42)
n = 100
x = np.random.normal(loc=1.78, scale=.1, size=n)

# t_value, p_value
t_value, p_value = stat.ttest_1samp(x,1.75)
print("t_value: %.4f, p_value: %.4f"%(t_value,p_value))

t_values = np.linspace(-5,5,100)
plt.plot(t_values,stat.t.pdf(t_values, n-1),'b-',label="T(n-1)")
upper = t_values[t_values > t_value]
plt.fill_between(upper, 0, stat.t.pdf(upper,n-1), label="p_value")


# two-sided
lower = t_values[t_values < -t_value]
plt.fill_between(lower, 0, stat.t.pdf(lower,n-1), label="lower side of p_value")

plt.legend()
plt.show()
```

## two-smaple t-test
### Task
1. Create two normally distributed random data sets, data_1 and data_2.

```python
np.random.seed(42)
data_1 = np.random.randn(1000)*4+100.5
data_2 = np.random.randn(1000)*3+101
```

### Task
1. t-test them. See if they can be considered equal.
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html

```python
t_value, p_value = stat.ttest_ind(data_1, data_2)
print('t_value = %.3f, p_value = %.3f'%(t_value, p_value))

alpha = 0.05
if p_value > alpha:
    print('They are equal (fail to reject H0)')
else:
    print('They are not equal (reject H0)')
```

### Task
1. Given the two samples, tell if their means are equal, using t-test, assuming equal variance. 

```python
# Given
np.random.seed(42)
n_x, n_y = 50, 25
x = np.random.normal(loc=1.76, scale=.1, size = n_x)
y = np.random.normal(loc=1.70, scale = .12, size = n_y)

# t_value, p_value
t_value, p_value = stat.ttest_ind(x,y,equal_var=True)
print("t_value: %.4f, p_value: %.4f"%(t_value,p_value))

# hypothesis test
alpha = 0.05
if p_value > alpha:
    print('They are equal (fail to reject H0)')
else:
    print('They are not equal (reject H0)')
```
